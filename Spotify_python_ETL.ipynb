```python
#refer spotipy documentation
#import required libraries
import pandas as pd
import os
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

#Accessing enviornment variables 
client_id = os.getenv("CLIENT_ID")
client_secret = os.getenv("CLIENT_SECRET")

#Athuntecation for spotify API
client_credentials = SpotifyClientCredentials(client_id =client_id, client_secret =client_secret)

#Authorization for spotify API
sp = spotipy.Spotify(auth_manager =client_credentials)

#URl link for the desired playlist in spotify. ex: here i have selected telugu love songs playlist from spotify
playlist_link = "https://open.spotify.com/playlist/37i9dQZF1DX44F1QWqYoaV"

#spliting URl link to get the URI to acess that playlist
playlist_URI = playlist_link.split("/")[4]

#storing the json data obtained from the above URI in a variable 
raw_data = sp.playlist_tracks(playlist_URI)

#Transforming the raw json data into list of dict data containg required fields
#making three lists album, artist and songs
#Obtaining the required fields in albums list from json file and storing those dictinories in list
album_list = []
for row in raw_data["items"]:
    album_id = row["track"]["album"]["id"]
    album_name = row["track"]["album"]["name"].split(" ")[0]
    album_release_date = row["track"]["album"]['release_date']
    album_url = row["track"]["album"]['external_urls']['spotify']
    album_elements = {"album_id" : album_id, "album_name" : album_name, "album_release_date" : album_release_date, "album_url" : album_url}
    album_list.append(album_elements)

#Obtaining the required fields in artist list from json file and storing those dictinories in list. 
#Here we have 1 or more artists for single album, iterating through artist and storing there data
artist_list = []
for row in raw_data["items"]:
    for key,value in row.items():
        if key == 'track':
            for artist in value['artists']:
                artist_dict = {'artist_id':artist['id'], 'artist_name':artist['name'], 'external_url': artist['external_urls']['spotify']}
                artist_list.append(artist_dict)

#Obtaining the required fields in song list from json file and storing those dictinories in list.
songs_list = []
for row in raw_data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,
                    'popularity':song_popularity,'song_added':song_added,'album_id':album_id,
                    'artist_id':artist_id
                   }
    songs_list.append(song_element)

#converting those dict list in to proper table format using pandas and deleting duplicate values
album_df = pd.DataFrame.from_dict(album_list)
album_df.drop_duplicates(subset = "album_id")

artist_df = pd.DataFrame.from_dict(artist_list)
artist_df.drop_duplicates(subset = "artist_id")

songs_df = pd.DataFrame.from_dict(songs_list)
songs_df.drop_duplicates(subset = "song_id")



```
